```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
lr_min = 0.001
lr_max = 1.0
```

```
step_size = 30 # half a triangle
num_cycles = 5 # num of triangles
```

```

total_iters = step_size * 2 * num_cycles
```

```
def calc_lr(iter):
  cycle = np.floor(1+ iter/(2*step_size))
  x = np.abs(iter/step_size - (2*cycle) + 1)
  lr_t = lr_min + (lr_max - lr_min)*(1-x)
  return lr_t
```

```
print(total_iters)
```

```
x = []
y = []
for i in np.linspace(0, total_iters, 500):
  x.append(i)
  y.append(calc_lr(i))
```

```
# print(x)
```

```
plt.figure(figsize=(12,6))
plt.plot(x, y)
plt.show()
```
